```python
data_shuffle=data.sample(frac=1)
data_shuffle=data_shuffle.reset_index(drop=True)
```

```python
x=data_shuffle.drop("Attrition",axis=1)
y=data_shuffle["Attrition"]
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler,OneHotEncoder
```

```python
CT=make_column_transformer(
(MinMaxScaler(),['Age','DailyRate','DistanceFromHome','Education','EnvironmentSatisfaction','HourlyRate','JobInvolvement',
'JobLevel','JobSatisfaction','MonthlyIncome','MonthlyRate','NumCompaniesWorked','OverTime',
'PercentSalaryHike','PerformanceRating','RelationshipSatisfaction','StockOptionLevel','TotalWorkingYears','TrainingTimesLastYear'
,'WorkLifeBalance','YearsAtCompany','YearsInCurrentRole','YearsSinceLastPromotion','YearsWithCurrManager']),
 (OneHotEncoder(),["BusinessTravel",'Department','EducationField','Gender','JobRole','MaritalStatus'])
)
```

```python
CT.fit(x)
x_scaled=CT.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_all_train,x_test,y_all_train,y_test=train_test_split(x_scaled,y,test_size=0.1,random_state=365)
```

```python
x_train,x_validation,y_train,y_validation=train_test_split(x_all_train,y_all_train,test_size=0.1,random_state=365)
```

```python
x_train.shape,x_validation.shape,x_test.shape,y_train.shape,y_validation.shape,y_test.shape
```

```python
import tensorflow as tf
```

```python
train_inputs=tf.constant(x_train)
train_targets=tf.constant(y_train)

validation_inputs=tf.constant(x_validation)
validation_targets=tf.constant(y_validation)

test_inputs=tf.constant(x_test)
test_targets=tf.constant(y_test)
```

```python
early_stopping=tf.keras.callbacks.EarlyStopping(patience=5)


model=tf.keras.Sequential([
                     tf.keras.layers.Dense(100, activation="relu"),
                     tf.keras.layers.Dense(100, activation="tanh"),
                     tf.keras.layers.Dense(100, activation="sigmoid"),
                     tf.keras.layers.Dense(100,activation="softmax")
                    ])

model.compile(optimizer ="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])

model.fit(train_inputs,train_targets,batch_size=50,callbacks=[early_stopping],
          validation_data=(validation_inputs, validation_targets),epochs=100, verbose=2)
```

```python
test_loss,test_accuracy=model.evaluate(test_inputs,test_targets)
```

```python
predict_targets=model.predict(test_inputs)
```

```python
predict_targets=predict_targets.argmax(axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
LogModel=LogisticRegression()
```

```python
x_log_train,x_log_test,y_log_train,y_log_test=train_test_split(x_scaled,y,test_size=0.2,random_state=365)
```

```python
LogModel.fit(x_log_train,y_log_train)
```

```python
y_log_predict=LogModel.predict(x_log_test)
```

```python
print("Neural Network Model Accuracy Rate{:2f}%".format(100*accuracy_score(test_targets,predict_targets)))
```

```python
print("Logistic Model Accuracy Rate{:2f}%".format(100*accuracy_score(y_log_test, y_log_predict)))
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(predict_targets,test_targets))
```

```python
print(classification_report(y_log_predict,y_log_test))
```
